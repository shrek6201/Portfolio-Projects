```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) #adjusted the configuration of the plots we will create

#read in the data

df = pd.read_csv(r'E:\Portfolio Projects\movies.csv')
```

```python
#looking at the data

df.head()
```

```python
#checking for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#data types for columns

df.dtypes
```

```python
print(df['budget'].isna().sum())
print(df['gross'].isna().sum())
```

```python
print(np.isinf(df['budget']).sum())
print(np.isinf(df['gross']).sum())
```

```python
df['budget'] = df['budget'].fillna(0).replace([np.inf, -np.inf], 0)
df['gross'] = df['gross'].fillna(0).replace([np.inf, -np.inf], 0)
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.dropna()
```

```python
#created correct year column

df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#drop any duplicates
df.drop_duplicates()
```

```python
df.head()
```

```python
#budget high correlation
#company high correlations
```

```python
#scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```
